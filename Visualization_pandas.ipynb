Visualization of data is important to understand the nuances of your dataset. Before we start Pandas Virtualization, we have to import the essential libraries. NumPy, Pandas, and Matplotlib.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dataflair = pd.read_csv("https://download.mlcc.google.com/mledu-datasets/california_housing_train.csv", sep=",")
dataflair.head()
```

Histogram

With the help of .hist() function, we can plot a histogram based on the parameters.

```python
dataflair.hist(column='households')
plt.show()
```

**Line chart**

The plot.line function gives a line plot. We can set the x and y axis.

```python
dataflair.plot.line(x='population', y='median_income', figsize=(8,6))
plt.show()
```

**Scatter plot**

The .scatter function lets us plot a scatter graph. Just like the previous function, the x and y-axes can be defined and the size of the graph can be set by the figsize parameter.

```python
dataflair.plot.scatter(x='population', y='median_income')
plt.show()
```

**Boxplot**

A boxplot is basically a five number summary of the data. It consists of the minimum, maximum, first quartile, median or second quartile, and the third quartile. The small dots are the outliers of the data. Here, we use .box function to plot a box plot graph.Boxplot can be drawn calling Series.plot.box() and DataFrame.box.plot(), or DataFrame.boxplot() to visualize the distribution of values within each column.

```python
dataflair.plot.box(figsize=(8,6))
plt.show()
```

**Hexagonal chart**

In such a graph a hexagon represents points of intersection. The increase in the points of intersection increases the darkness of the color of the hexagon. We can create a hexagonal chart with the help of .plot.hexbin function.

```python
dataflair.plot.hexbin(x='population', y='median_income',gridsize=30)
plt.show()
```

**Pie chart**

We can plot a pie chart with the help of .plot.pie function.

```python
dataflair2= pd.DataFrame({'cost': [79, 40 , 60]},index=['Oranges', 'Bananas', 'Apples'])
dataflair2.plot.pie(y='cost', figsize=(8, 6))
plt.show()
```

**Kernel density chart**

We can plot kernel density graphs with the help of the .kde function. This gives us the graph for a particular column.

```python
dataflair["median_income"].plot.kde()
plt.show()
```
